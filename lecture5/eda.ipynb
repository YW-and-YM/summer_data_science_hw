# Exploratory data analysis

First, load the dataset

```python
import pandas as pd
```

```python
salary_data = pd.read_csv("../lecture1/Salary.csv")
salary_data
```

## 1. What is the median salary
### 1.1 Median salary of the entire dataset

```python
average = salary_data["Salary"].mean()
median = salary_data["Salary"].median()
print("Average salary: ", average)
print("Median salary: ", median)
```

### 1.2 Median salary of each education level

```python
median_salary_per_education_level = {"Education Level": [], "Median Salary": []}
levels = salary_data["Education Level"].unique()
for level in levels:
    median_salary_per_education_level["Education Level"].append(level)
    median_salary_per_education_level["Median Salary"].append(salary_data[salary_data["Education Level"] == level]["Salary"].median())

pd.DataFrame(median_salary_per_education_level).sort_values(by="Median Salary")
```
