## Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_path = "/Users/yuema/MyCode/summer_data_science_hw/lecture3/Salary.csv"
salary_data = pd.read_csv(data_path)
```

```python
salary_data
```

```python
salary_threshold = np.median(salary_data["Salary"])
```

```python
salary_data["Salary"]
```

```python
salary_threshold
```

```python
salary_level_list = []
for i in range(len(salary_data)):
    if salary_data.iloc[i]["Salary"] < salary_threshold:
        salary_level_list.append("Low")
    else:
        salary_level_list.append("High")
```

```python
salary_data["salary_level"] = np.array(salary_level_list)
```

```python
salary_data
```

```python
mask_level_0 = salary_data["Education Level"] == 0
people_level_0 = salary_data[mask_level_0]
```

```python
mask_level_0
```

```python
people_level_0
```

```python
plt.hist(people_level_0["salary_level"])
plt.show()
```

```python
mask_level_1 = salary_data["Education Level"] == 1
people_level_1 = salary_data[mask_level_1]
plt.hist(people_level_1["salary_level"])
plt.show()
```

```python
mask_level_2 = salary_data["Education Level"] == 2
people_level_2 = salary_data[mask_level_2]
plt.hist(people_level_2["salary_level"])
plt.show()
```

### Let's do a p-value test for education level 1

```python
total_people = len(people_level_1)
low_salary_people = len(people_level_1[people_level_1["salary_level"] == "Low"])
high_salary_people = len(people_level_1[people_level_1["salary_level"] == "High"])
print("Total people in education level 1 is: ", total_people)
print("The number of low salary people in education level 1 is: ", low_salary_people)
print("The number of high salary people in education level 1 is: ", high_salary_people)
print("The percentage of high salary people in education level 1 is: ", high_salary_people / total_people)
```

High Salary <-> Head of the coin
