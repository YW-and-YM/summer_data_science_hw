## Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_path = "/Users/yuema/MyCode/summer_data_science_hw/lecture3/Salary.csv"
salary_data = pd.read_csv(data_path)
```

```python
salary_data
```

```python
salary_threshold = np.median(salary_data["Salary"])
```

```python
salary_data["Salary"]
```

```python
salary_threshold
```

```python
salary_level_list = []
for i in range(len(salary_data)):
    if salary_data.iloc[i]["Salary"] < salary_threshold:
        salary_level_list.append("Low")
    else:
        salary_level_list.append("High")
```

```python
salary_data["salary_level"] = np.array(salary_level_list)
```

```python
salary_data
```

```python
mask_level_0 = salary_data["Education Level"] == 0
people_level_0 = salary_data[mask_level_0]
```

```python
plt.hist(people_level_0["salary_level"])
plt.show()
```
